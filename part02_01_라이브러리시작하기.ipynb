```python
x = np.linspace(0,14,100) # linspace (시작값, 끝값, 개수)
x
```

```python
y = np.sin(x) # 100개의 x의 값에 대해서 100개의 sin값을 구해서 y로 넣어준다.
plt.plot(x,y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x,y1, x, y2, x, y3)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='sin(x)*2')
plt.plot(x,y3, label='sin(x)*4')
plt.legend()
plt.title("sin plot")
```

## 실습 4-1) cos 그래프 그려보기 3개

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4

# 그래프 그리기
plt.plot(x,y1, label='cos(x)')
plt.plot(x,y2, label='cos(x)*2')
plt.plot(x,y3, label='cos(x)*4')
plt.legend()
plt.title("cos plot")
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x,y1, 'o' , label='sin(x)')  # 'o' 는 점으로 찍어주는 것
plt.plot(x,y2, 'v' , label='sin(x)*2')  # 'v' 는 삼각형
plt.plot(x,y3, '^' , label='sin(x)*4')    # '^'는 역삼각형
plt.legend()
plt.title("sin plot")
```

# 요일별 식사 금액은 얼마나 될까?

```python
import seaborn as sns
print(sns.__version__)
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
# 제목 Amount of meal
plt.title('Amount of meal')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
# 제목 Amount of meal
plt.title('Amount of meal')
```

```python
# Initialize a grid of plots tip an Axes for each walk
grid = sns.FacetGrid(tips, col="size", hue="size", palette="tab20c",
                     col_wrap=4, height=1.5)

# Draw a horizontal line to show the starting point
grid.map(plt.axhline, y=0, ls=":", c=".5")

# Draw a line plot to show the trajectory of each random walk
grid.map(plt.plot, "tip", "total_bill", marker="o")

# Adjust the tick positions and labels
grid.set(xticks=np.arange(5), yticks=[-3, 3],
         xlim=(-.5, 4.5), ylim=(-3.5, 3.5))
```
